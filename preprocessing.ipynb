```python
import pandas as pd
import numpy as np
from datetime import date
import re

import spacy
print(spacy.prefer_gpu())
spacy.prefer_gpu(gpu_id=0)

import os
import re
# from tqdm.notebook import tqdm
from tqdm import tqdm
tqdm.pandas()
import pickle

from flashtext import KeywordProcessor
kp = KeywordProcessor()

from spacy.lang.en.stop_words import STOP_WORDS
```

```python
nlp = spacy.load('en_core_web_md', disable=['parser', 'ner']) # 정확도 향상을 위해 md 모델 사용
# nlp = spacy.load('en_core_web_trf', disable=['parser', 'ner']) # transformer 기반 trf 모델

# spacy의 불용어 리스트를 stop_words 변수에 할당합니다.
stop_words = STOP_WORDS

# Define functions for stopwords, bigrams, trigrams and lemmatization
def remove_stopwords(texts):
    processed_texts = []
    for doc in texts:
        # 문자열을 소문자로 변환하고, 정규식으로 단어 단위(token) 추출
        tokens = re.findall(r'\b\w+\b', str(doc).lower())
        # stop_words에 포함되지 않은 단어만 필터링
        filtered_tokens = [token for token in tokens if token not in stop_words]
        processed_texts.append(filtered_tokens)
    return processed_texts

# def make_bigrams(texts):
#         return [bigram_mod[doc] for doc in texts]

# def make_trigrams(texts):
#        return [trigram_mod[bigram_mod[doc]] for doc in texts]

def lemmatization(texts, allowed_postags=None): #https://spacy.io/api/annotation#
        texts_out = []
        print('lemmatization 시작 \n')
        for sent in tqdm(texts):
                doc = nlp(" ".join(sent))
                if allowed_postags != None:
                        texts_out.append([token.lemma_ for token in doc if token.pos_ in allowed_postags])
                else:
                        texts_out.append([token.lemma_ for token in doc])
        return texts_out

def tokenize_for_topicmodel(data_words):
    # 일반명사, 고유명사, 형용사, 부사, 동사 선택하여 표제어 추출
    data_words_lemmatized = lemmatization(data_words, allowed_postags=['NOUN','PROPN']) # ,'ADJ','ADV','VERB'
    # lemmatization 결과에서 불용어 제거
    data_words_filtered = remove_stopwords(data_words_lemmatized)
    return data_words_filtered
```

```python
# 토크나이징 test
tokenize_for_topicmodel([['wear a masks'],['coverings facecoverings facecovers joe biden'],['republicans republican']])
```

```python
# filtered_df의 "Abstract" 열은 각 문서가 문자열로 되어 있다고 가정합니다.
# 각 문자열을 리스트로 감싸서 tokenize_for_topicmodel 함수의 입력 형식에 맞춥니다.
abstract_list = filtered_df['Abstract'].apply(lambda x: [str(x)]).tolist()

# tokenize_for_topicmodel 함수 적용
tokenized_abstracts = tokenize_for_topicmodel(abstract_list)

# 결과를 "Tokenized_Abstract"라는 새 열로 DataFrame에 추가합니다.
filtered_df['Tokenized_Abstract'] = tokenized_abstracts

# 확인: 새 열의 일부 결과 출력
print(filtered_df[['Abstract', 'Tokenized_Abstract']].head())
```

```python
# Tokenized_Abstract 리스트를 문자열로 변환하여 새로운 열 추가
filtered_df['Preprocessed_Abstract'] = filtered_df['Tokenized_Abstract'].apply(lambda tokens: ' '.join(tokens))

# 결과 확인
print(filtered_df[['Tokenized_Abstract', 'Preprocessed_Abstract']].head())
```
